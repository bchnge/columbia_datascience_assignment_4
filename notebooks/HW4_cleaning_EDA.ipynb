```python
# Load in the small fec data.  You should have this data (fec_small.h5)
stpath = '../data/raw/fec_small.h5'
store = pd.HDFStore(stpath)
frame = store.select('fec_small')
store.close()
```

Make the columns a little easier to read

```python
frame.columns = ['Candidate', 'Contributor', 
                 'City', 'State', 'Zipcode',
                 'Employer', 'Occupation',
                 'Amount', 'Date']
frame
```

Add party affiliations for each candidate using merge

```python
party = pd.read_csv('../data/raw/cand_af.csv', index_col=0,
 header=None, names=['Candidate', 'PoliticalParty'])
party
```

```python
# Join contributions with the party affiliation
joined = pd.merge(frame, party, left_on='Candidate', right_index=True)
```

### Time to Party

```python
rs = joined.groupby('PoliticalParty').sum().sort_index()
rs # notice how pandas excludes nuisance columns to start
```

```python
rs.plot(kind='bar') # Y-axis in units of $10mm
# rs.plot(logy=True)
```

### States

```python
# Load in state FIPS codes  http://www.columbia.edu/~sue/state-fips.html
states = pd.read_csv('../data/raw/state_id.csv', index_col=0, header=None, 
                    names=['State', 'StateID'])
states.tail(10)
```

## Cleaning

---

```python
frame2 = joined.copy()
```

```python
# Replace some categories with N/A
frame2.Employer.replace(['REFUSED', 'REQUESTED', 'INFORMATION REQUESTED', 'INFORMATION REQUESTED PER BEST EFFORTS'], 'N/A', inplace=True)
frame2.Occupation.replace(['REFUSED', 'REQUESTED', 'INFORMATION REQUESTED', 'INFORMATION REQUESTED PER BEST EFFORTS'], 'N/A', inplace=True);
```

```python
frame2.Employer.replace(['SELF-EMPLOYED', 'SELF'], 'SELF EMPLOYED', inplace=True)
```

```python
by_employer_clean = frame2.groupby('Employer').sum().sort('Amount', ascending=False)
```

```python
by_employer_clean.drop('StateID', axis=1).head(10).plot(kind='barh')
```

```python
func = lambda x: x.groupby('Employer').sum().sort('Amount', ascending=False).head(10)
```

```python
result = frame2.groupby('PoliticalParty').apply(func)
```
